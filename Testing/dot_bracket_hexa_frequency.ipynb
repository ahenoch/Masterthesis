```python
from matplotlib import pyplot as plt    
from matplotlib import style    
```

```python
def convert_graph(bracket):
    notation = []
    depth = 0
    for letter in bracket:
        if letter == '(':
            notation += [depth]
            depth += 1
        elif letter == ')':
            depth -= 1
            notation += [depth]
        elif letter == '.':
            notation += [depth]
        else:
            print('Error!')
            break
    return(notation)
```

```python
style.use('ggplot')      

y1 = convert_graph(one)
y2 = convert_graph(two)
y3 = convert_graph(three)
y4 = convert_graph(four)

x1 = list(range(len(y1)))    
x2 = list(range(len(y2)))
x3 = list(range(len(y3)))    
x4 = list(range(len(y4)))  

plt.plot(x1, y1, 'blue', label='cluster 1', linewidth=1)    
plt.plot(x2, y2, 'red', label='cluster 2', linewidth=1)
plt.plot(x3, y3, 'green', label='cluster 3', linewidth=1)    
plt.plot(x4, y4, 'black', label='cluster 4', linewidth=1)

plt.title('')    
plt.ylabel('Pairing Depth')    
plt.xlabel('Position') 
plt.legend()
plt.show()  
```

```python
def convert(bracket):
    notation = []
    depth = 2
    for letter in bracket:
        if depth == 0:
            print('Error!')
            break
        elif depth > 248:
            print('Error!')
        elif letter == '(':
            while hex(depth)[2:].zfill(2) in ['3e', '3d', '3c', '2d', '20', '0d', '0a']:
                depth += 1
            notation += [hex(depth)[2:].zfill(2)]
            depth += 1
        elif letter == ')':
            depth -= 1
            while hex(depth)[2:].zfill(2) in ['3e', '3d', '3c', '2d', '20', '0d', '0a']:
                depth -= 1
            notation += [hex(depth)[2:].zfill(2)]
        elif letter == '.':
            notation += [hex(1)[2:].zfill(2)]
        else:
            print('Error!')
            break
    return(" ".join(notation))
```

```python
def convert_back(notation):
    bracket = ""
    depth = 1
    for hexa in notation:
        if depth == 0:
            print('Error!')
            break
        elif hexa == "--":
            bracket += "-"
        elif int(hexa, 16) == 1:
            bracket += "."
        elif int(hexa, 16) > depth:
            bracket += "("
            depth += 1
        elif int(hexa, 16) <= depth:
            bracket += ")"
            depth -= 1
        else:
            print('Error!')
            break
    return(bracket)
```
