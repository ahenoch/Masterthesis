```python
import re
import sys
import csv
import functools as ft
import numpy as np
import pandas as pd
import multiprocessing as mp
import collections as co
import itertools as it
import difflib as dl
import progressbar
```

```python
class readingframe(object):
    
    def __init__(self, worldfile, procs = 8):
    
        self.procs = procs
        self.cities = co.defaultdict(list)
        self.subcountries = co.defaultdict(list)
        self.countries = co.defaultdict(list)

        chunk = pd.read_csv(worldfile, chunksize = 10000, sep = ',', na_filter = False)
        for data in chunk:
        
            for city, country, subcountry, geonameid in data.itertuples(index=False, name=None):
                
                self.cities[city].append([city, subcountry, country])
                self.subcountries[subcountry]. append(['null', subcountry, country])
                self.countries[country].append(['null', 'null', country])
        
        self.cities_names = list(self.cities.keys())
        self.subcountries_names = list(self.subcountries.keys())
        self.countries_names = list(self.countries.keys())


    def search(self, entry):

        match_subcountry = dl.get_close_matches(entry, self.subcountries_names, 1, 0.9)
        if not match_subcountry:
            match_city = dl.get_close_matches(entry, self.cities_names, 1, 0.9)
            if not match_city:
                match_country = dl.get_close_matches(entry, self.countries_names, 1, 0.9)
                if not match_country:
                    result = ['null', 'null', 'null']
                else:
                    match = match_country[0]
                    result = self.countries[match]
            else:
                match = match_city[0]
                result = self.cities[match]
        else:
            match = match_subcountry[0]
            result = self.subcountries[match]

        if any(isinstance(i, list) for i in result):
            output = result[0]
        else:
            output = result

        return(output)


    def worker(self, line):

        line = list(map(str, line))

        head = line[0].split('|')

        acc = head[0][1:]
        name = head[1]
        seg = head[2]
        org = head[4]
        sub = head[5]
        if sub == 'NA' or sub == 'nan':
            sub = 'null'
        host = head[7]
        entry = line[1]

        info = name.split('/')

        spec = info[0]
        del info[0]
        year = info[-1]
        del info[-1]

        if year.isdecimal():
            if len(year) == 2:
                year = '19'+year 
        else:
            year = 'null'

        if not info:
            pos = ['null', 'null', 'null']
        else:
            for i in info:
                pos = self.search(i)
                if not all([item == 'null' for item in pos]):
                    break
                    
        out = entry#acc, name, seg, entry , name, spec, pos[0], pos[1], pos[2], year, sub, host
        
        return out


    def create_tables(self, infile, gnmsfile, strnsfile):

        chunk = pd.read_csv(infile, chunksize = 10000, sep = ';', na_filter = False, header = None)
        
        for data in chunk:

            with mp.Pool(self.procs) as pool:

                result = pool.imap(self.worker, data.itertuples(index=False, name=None), chunksize=20)
                result_final = np.concatenate(result)
        
            return(result_final)
                
                #with open(gnmsfile, 'a', newline='') as genomes, open(strnsfile, 'a', newline='') as strains:

                #    for out in result:

                #        out_gnms = csv.writer(genomes, delimiter="\t")
                #        out_gnms.writerow(out[0])

                #        out_strns = csv.writer(strains, delimiter="\t")
                #        out_strns.writerow(out[1])
```
